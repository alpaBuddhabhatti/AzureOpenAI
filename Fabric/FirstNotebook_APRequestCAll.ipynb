```python
import os
import base64
import requests

# Define the Azure OpenAI endpoint and API key

azure_endpoint = "https://azureopenai24demo.openai.azure.com/openai/deployments/gpt-35-turbo/chat/completions?api-version=2024-05-01-preview"
api_key = "7c15f7bb4139435bbefa4b8e155a1622"
api_version = "2024-05-01-preview"

# Replace <deployment-name> with your actual deployment name

# Prepare the chat prompt
chat_prompt = [
    {
        "role": "system",
        "content": "You are an AI assistant that helps people find information."
    }
]

# Create the request payload
payload = {
    "model": "gpt-35-turbo",
    "messages": chat_prompt,
    "temperature": 0.7,
    "max_tokens": 800,
    "top_p": 0.95,
    "frequency_penalty": 0,
    "presence_penalty": 0,
    "stop": None
}

# Define the headers
headers = {
    "api-key": api_key,
    "Content-Type": "application/json",
    "api-version": api_version
}

# Send the request using requests
response = requests.post(
    azure_endpoint,
    json=payload,
    headers=headers
)

# Check for errors
if response.status_code == 200:
    completion = response.json()
    content = completion['choices'][0]['message']['content']
    print(content)  # Print only the content
else:
    print(f"Error: {response.status_code} - {response.text}")
```
