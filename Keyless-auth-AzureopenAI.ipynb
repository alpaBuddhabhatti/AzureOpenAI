**Keyless auth - Azure OpenAI with Managed Identity**

```python
import os
from openai import AzureOpenAI
from azure.identity import DefaultAzureCredential
from azure.core.credentials import AzureKeyCredential


# Initialize the DefaultAzureCredential to use Managed Identity
credential = DefaultAzureCredential()

# Retrieve the endpoint URL from environment variables or use the default value
endpoint = os.getenv("ENDPOINT_URL", "https://YOUR-AZUREAI-INSTANT.openai.azure.com/")

# Set up Azure OpenAI client with Managed Identity authentication
client = AzureOpenAI(
    azure_endpoint=endpoint,
    azure_ad_token_provider=lambda: credential.get_token("https://cognitiveservices.azure.com/.default").token,
    api_version="2024-05-01-preview", #your Azure OpenAI VERSION 
)


# Create a completion request with Azure Search integration using Managed Identity
completion = client.chat.completions.create(
    model="GPT-40", # YOUR AZURE OPENAI MODEL NAME
    messages=[{
        "role": "user",
        "content": "What are the differences between Azure Machine Learning and Azure AI services?"
    }],
    max_tokens=800,
    temperature=0,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0,
    stop=None,
    stream=False
    )


message_content = completion.choices[0].message.content

# Print the content of the message
print(message_content)
```
